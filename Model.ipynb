```python
X.drop(['under-five deaths _t',' thinness 5-9 years_t','Polio_t','Schooling_t'],axis=1,inplace=True)
```

```python
X
```

```python
import statsmodels.api as sm
from sklearn import metrics
X=sm.add_constant(X)
model=sm.OLS(y,X)
results=model.fit()
print(results.summary())
```

```python
X.drop(['Status','Measles _t','Total expenditure_t'],axis=1,inplace=True)
```

```python
X.drop('const',axis=1,inplace=True)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import RobustScaler
rs=RobustScaler()
X_train=rs.fit_transform(X_train)
X_test=rs.transform(X_test)
```

```python
from sklearn.feature_selection import mutual_info_classif
importances=mutual_info_classif(X,y.astype('int'))
feat_importances=pd.Series(importances,X.columns[0:len(X.columns)])
feat_importances.plot(kind='barh',color='teal')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
reg=LinearRegression().fit(X_train,y_train)
pred=reg.predict(X_test)
r2_score(y_test,pred)
```

```python
mean_squared_error(y_test,pred)
```

```python
from sklearn.model_selection import cross_val_score
accuracies=cross_val_score(estimator=reg,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.svm import SVR
svr=SVR().fit(X_train,y_train)
pred=svr.predict(X_test)
r2_score(y_test,pred)
```

```python
mean_squared_error(y_test,pred)
```

```python
accuracies=cross_val_score(estimator=svr,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.tree import DecisionTreeRegressor
dec=DecisionTreeRegressor(random_state=0).fit(X_train,y_train)
pred=dec.predict(X_test)
r2_score(pred,y_test)
```

```python
mean_squared_error(y_test,pred)
```

```python
accuracies=cross_val_score(estimator=dec,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(random_state=0).fit(X_train,y_train)
pred=rf.predict(X_test)
r2_score(pred,y_test)
```

```python
mean_squared_error(y_test,pred)
```

```python
accuracies=cross_val_score(estimator=rf,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```
