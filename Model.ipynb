```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('transformed.csv')
data
```

```python
data.isnull().sum()
```

```python
X=data.iloc[:,np.r_[2:4,22:39]]
```

```python
X
```

```python
X.columns.to_list()
```

```python
y=data['Life expectancy ']
```

```python
y
```

```python
correlation_matrix=X.corr()
plt.figure(figsize=(20,20))
sns.heatmap(correlation_matrix , cmap='RdBu_r', annot=True)
plt.show()
```

```python
correlated_features =[]
for i in range(len(correlation_matrix .columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > 0.8:
            colname = [correlation_matrix.columns[i],correlation_matrix.columns[j]]
            correlated_features.append(colname)
print(correlated_features)
```

```python
X.drop(['under-five deaths _t',' thinness 5-9 years_t','Polio_t','Schooling_t'],axis=1,inplace=True)
```

```python
X
```

```python
import statsmodels.api as sm
from sklearn import metrics
X=sm.add_constant(X)
model=sm.OLS(y,X)
results=model.fit()
print(results.summary())
```

```python
X.drop(['Status','Measles _t','Total expenditure_t'],axis=1,inplace=True)
```

```python
X.drop('const',axis=1,inplace=True)
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import RobustScaler
rs=RobustScaler()
X_train=rs.fit_transform(X_train)
X_test=rs.transform(X_test)
```

```python
from sklearn.feature_selection import mutual_info_classif
importances=mutual_info_classif(X,y.astype('int'))
feat_importances=pd.Series(importances,X.columns[0:len(X.columns)])
feat_importances.plot(kind='barh',color='teal')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
reg=LinearRegression().fit(X_train,y_train)
pred=reg.predict(X_test)
r2_score(y_test,pred)
```

```python
mean_squared_error(y_test,pred)
```

```python
from sklearn.model_selection import cross_val_score
accuracies=cross_val_score(estimator=reg,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.svm import SVR
svr=SVR().fit(X_train,y_train)
pred=svr.predict(X_test)
r2_score(y_test,pred)
```

```python
mean_squared_error(y_test,pred)
```

```python
accuracies=cross_val_score(estimator=svr,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.tree import DecisionTreeRegressor
dec=DecisionTreeRegressor(random_state=0).fit(X_train,y_train)
pred=dec.predict(X_test)
r2_score(pred,y_test)
```

```python
mean_squared_error(y_test,pred)
```

```python
accuracies=cross_val_score(estimator=dec,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(random_state=0).fit(X_train,y_train)
pred=rf.predict(X_test)
r2_score(pred,y_test)
```

```python
mean_squared_error(y_test,pred)
```

```python
accuracies=cross_val_score(estimator=rf,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from xgboost import XGBRegressor
xgb_model = XGBRegressor(random_state=0)
xgb_model.fit(X_train, y_train)
pred = xgb_model.predict(X_test)
r2_score(y_test, pred)
```

```python
mean_squared_error(y_test,pred)
```

```python
accuracies=cross_val_score(estimator=xgb_model,X=X_train,y=y_train,cv=10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.model_selection import GridSearchCV
rf_model = RandomForestRegressor()
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, scoring='r2', cv=3)
grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
pred = best_model.predict(X_test)
r2 = r2_score(y_test, pred)
print("Best Hyperparameters:", best_params)
print("R-squared Score on Test Set:", r2)
```

```python
xgb_model = XGBRegressor()
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, scoring='r2', cv=3)
grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
pred = best_model.predict(X_test)
r2 = r2_score(y_test, pred)

print("Best Hyperparameters:", best_params)
print("R-squared Score on Test Set:", r2)
```

```python
from xgboost import XGBRegressor
xgb_model = XGBRegressor(random_state=0,learning_rate=0.2, max_depth=5, n_estimators=200)
xgb_model.fit(X_train, y_train)
pred = xgb_model.predict(X_test)
r2_score(y_test, pred)
```
